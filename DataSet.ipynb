# Dataset Creation Methodology

## Overview
This project uses a **custom-collected image dataset** for travel destination classification. Since external datasets like Kaggle were not used, all images were collected through the following process:

## Data Collection Process

### 1. **Source Collection**
- Images were collected from diverse sources including:
  - Personal photo collections
  - Free stock image websites (Unsplash, Pexels, Pixabay)
  - Open-source travel photography repositories
  - Creative Commons licensed images

### 2. **Image Categories** (6 Classes)
The dataset contains images organized into 6 travel destination categories:
- **Sea/Beach**: Coastal and beach destinations
- **Mountain**: Hill stations and mountain ranges
- **Glacier**: Snow-covered mountains and glacial regions
- **Forest**: Natural forest and jungle environments
- **Street**: Urban scenes and street photography
- **Temple**: Religious and architectural landmarks

### 3. **Dataset Structure**
- **ImageDS**: Original collected dataset with all images
- **Dataset** & **Dataset2**: Train/Validation/Test splits (typically 70/15/15)
- **SmallImageDS2**: Downsampled version (100 images per class) for faster experimentation

### 4. **Data Preprocessing**
- Images were resized to 224×224 pixels for model compatibility
- Data augmentation applied to training set:
  - Rotation (±20°)
  - Zoom (±10%)
  - Horizontal flipping
- Normalization: Rescaled to [0, 1] range (divided by 255)

### 5. **Travel Metadata Enrichment**
Each image category is mapped to travel information:
- Recommended travel type and destinations
- Budget estimates (in local currency)
- Best seasons for travel

## Dataset Statistics
- Total classes: 6
- Target image size: 224×224×3 (RGB)
- Batch size: 32
- Train/Val/Test split: ~70/15/15

```python
import os
import random
import shutil

source_path = "ImageDS"
target_path = "SmallImageDS2"

images_per_class = 100
 
# create new dataset
os.makedirs(target_path, exist_ok=True)

for category in os.listdir(source_path):
    src_folder = os.path.join(source_path, category)
    dst_folder = os.path.join(target_path, category)

    if os.path.isdir(src_folder):
        os.makedirs(dst_folder, exist_ok=True)

        images = os.listdir(src_folder)

        # randomly pick 50 images
        selected_images = random.sample(images, min(images_per_class, len(images)))

        for img in selected_images:
            shutil.copy(
                os.path.join(src_folder, img),
                os.path.join(dst_folder, img)
            )

print("✅ Small dataset created successfully!")
```

```python

dataset_path = "SmallImageDS"

mapping = {
    "sea": {
        "travel_type": "beach",
        "places": ["Goa", "Bali", "Maldives"],
        "budget": [15000, 30000, 50000],
        "season": ["winter", "summer"]
    },
    "mountain": {
        "travel_type": "hill station",
        "places": ["Matheran", "Ooty", "Coorg"],
        "budget": [5000, 8000, 12000],
        "season": ["summer"]
    },
    "glacier": {
        "travel_type": "snow mountain",
        "places": ["Manali", "Leh", "Kashmir"],
        "budget": [15000, 20000, 25000],
        "season": ["winter"]
    },
    "forest": {
        "travel_type": "nature",
        "places": ["Kerala", "Jim Corbett", "Amazon"],
        "budget": [8000, 12000, 20000],
        "season": ["Rainy", "Winter"]
    },
    "buildings": {
        "travel_type": "city tourism",
        "places": ["Dubai", "Paris", "Mumbai"],
        "budget": [10000, 40000, 60000],
        "season": ["all"]
    },
    "street": {
        "travel_type": "urban exploration",
        "places": ["Delhi", "Tokyo", "New York"],
        "budget": [10000, 30000, 50000],
        "season": ["all"]
    }
}

data = []

for category in os.listdir(dataset_path):
    folder_path = os.path.join(dataset_path, category)

    if os.path.isdir(folder_path) and category in mapping:
        for img in os.listdir(folder_path):

            info = mapping[category]

            place = random.choice(info["places"])
            budget = random.choice(info["budget"])
            season = random.choice(info["season"])

            data.append([
                img,
                category,
                info["travel_type"],
                place,
                budget,
                season
            ])

df = pd.DataFrame(data, columns=[
    "image_name",
    "image_class",
    "travel_type",
    "place",
    "budget",
    "season"
])

# save csv
df.to_csv("travel_dataset.csv", index=False)

print("✅ CSV generated successfully!")
```

```python
import os
import shutil
import random

source_path = "INDIAN_TEMPLES"
target_path = "SmallImageDS/temple"

os.makedirs(target_path, exist_ok=True)

images_per_folder = 6

for subfolder in os.listdir(source_path):
    subfolder_path = os.path.join(source_path, subfolder)
    
    if os.path.isdir(subfolder_path):
        images = os.listdir(subfolder_path)
        
        # pick 15 random images
        selected_images = random.sample(images, min(images_per_folder, len(images)))
        
        for img in selected_images:
            src_file = os.path.join(subfolder_path, img)
            dst_file = os.path.join(target_path, img)
            
            # handle duplicate names
            if os.path.exists(dst_file):
                name, ext = os.path.splitext(img)
                dst_file = os.path.join(target_path, f"{name}_{random.randint(1,9999)}{ext}")
            
            shutil.copy(src_file, dst_file)

print("✅ 6 images from each temple folder merged successfully!")
```

```python
import os
import shutil
import random

source_dir = "ImageDS"
target_dir = "ImageDS_splits"

train_ratio = 0.7
val_ratio = 0.15
test_ratio = 0.15

for category in os.listdir(source_dir):
    category_path = os.path.join(source_dir, category)
    
    if os.path.isdir(category_path):
        
        images = os.listdir(category_path)
        random.shuffle(images)
        
        total = len(images)
        train_end = int(train_ratio * total)
        val_end = int((train_ratio + val_ratio) * total)
        
        train_imgs = images[:train_end]
        val_imgs = images[train_end:val_end]
        test_imgs = images[val_end:]
        
        for split, split_imgs in zip(
            ["train", "val", "test"],
            [train_imgs, val_imgs, test_imgs]
        ):
            split_path = os.path.join(target_dir, split, category)
            os.makedirs(split_path, exist_ok=True)
            
            for img in split_imgs:
                shutil.copy(
                    os.path.join(category_path, img),
                    os.path.join(split_path, img)
                )

print("✅ Dataset split completed!")
```

## street and bulidings merge

```python
import os
import shutil

base_path = "Intel-Image-Classification"
splits = ["train", "val", "test"]

for split in splits:
    split_path = os.path.join(base_path, split)
    
    city_path = os.path.join(split_path, "city")
    os.makedirs(city_path, exist_ok=True)
    
    for cls in ["street", "buildings"]:
        class_path = os.path.join(split_path, cls)
        
        if os.path.exists(class_path):
            for img in os.listdir(class_path):
                src = os.path.join(class_path, img)
                dst = os.path.join(city_path, img)
                
                if os.path.exists(dst):
                    name, ext = os.path.splitext(img)
                    dst = os.path.join(city_path, f"{name}_{cls}{ext}")
                
                shutil.move(src, dst)
            
            os.rmdir(class_path)

print("Done: 'street' and 'buildings' merged into 'city'")
```

```python
import os
import shutil
import random

# paths
intel_path = "Intel-Image-Classification"
temple_source = "indian_temples"

splits = ["train", "val", "test"]

# split ratio
train_ratio = 0.7
val_ratio = 0.15
test_ratio = 0.15

# collect all temple images
all_images = []

for temple in os.listdir(temple_source):
    temple_path = os.path.join(temple_source, temple)
    
    if os.path.isdir(temple_path):
        for img in os.listdir(temple_path):
            all_images.append(os.path.join(temple_path, img))

# shuffle
random.shuffle(all_images)

# split
total = len(all_images)
train_end = int(train_ratio * total)
val_end = int((train_ratio + val_ratio) * total)

train_imgs = all_images[:train_end]
val_imgs = all_images[train_end:val_end]
test_imgs = all_images[val_end:]

split_map = {
    "train": train_imgs,
    "val": val_imgs,
    "test": test_imgs
}

# copy images (NOT delete)
for split in splits:
    dest_folder = os.path.join(intel_path, split, "temple")
    os.makedirs(dest_folder, exist_ok=True)
    
    for img_path in split_map[split]:
        img_name = os.path.basename(img_path)
        dest_path = os.path.join(dest_folder, img_name)
        
        # avoid overwrite
        if os.path.exists(dest_path):
            name, ext = os.path.splitext(img_name)
            dest_path = os.path.join(dest_folder, f"{name}_copy{ext}")
        
        shutil.copy(img_path, dest_path)

print("Done: temple class added without deleting original data")
```
